### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

```python
import numpy as np
import pandas as pd
```

### Series from lists

```python
# string
country = ['Nepal', 'India', 'Pakistan', 'Bhutan', 'Bangladesh']
pd.Series(country)
```

```python
# integers
runs = [13,24,56,78,100]
runs_series=pd.Series(runs)
```

```python
# custom index
marks = [67,57,89,100]
subjects = ['Math', 'English', 'Science', 'Nepali']
pd.Series(marks, index=subjects)
```

```python
# setting a name
pd.Series(marks, index=subjects, name='Marks')
```

### Series from dict

```python
marks = {
    'Math': 67,
    'English': 57,
    'Science': 89,
    'Nepali': 100
}
marks_series = pd.Series(marks, name = 'Marks')
marks_series
```
